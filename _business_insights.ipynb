# ** 1.What are the best locations to open new restaurants?**

```python
#Location with high ratings
import pandas as pd

df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv")
top_rated_locations=df.groupby('location')['rating'].mean().sort_values(ascending=False).head(5)
print(top_rated_locations)
```

```python
#Location with most votes
top_voted_locations=df.groupby('location')['votes'].sum().sort_values(ascending=False).head(5)
print(top_voted_locations)
```

```python
# Locations with High Demand but Fewer Restaurants
location_counts=df['location'].value_counts()
high_demand_low_supply=location_counts[location_counts < location_counts.mean()].index

top_high_demand=df[df['location'].isin(high_demand_low_supply)].groupby('location')['votes'].sum().sort_values(ascending=False).head(5)
print(top_high_demand)
```

## **Best Locations to Open a New Restaurant**

Based on **average ratings, total votes, and customer reviews**, the top locations are:

### **🏆 Top 5 Locations**
1️⃣ **Koramangala 5th Block** – Highest votes, strong customer base  
2️⃣ **Church Street** – High reviews & engagement  
3️⃣ **Indiranagar** – Popular & high customer interaction  
4️⃣ **Lavelle Road** – Best average ratings  
5️⃣ **St. Marks Road** – Strong ratings & customer feedback  

### **📌 Insights & Recommendations**
✔️ **High engagement areas attract more customers**  
✔️ **Focus on quality food & service to maintain ratings**  
✔️ **Leverage social media & customer loyalty programs**  

Opening a restaurant in these areas increases **visibility, engagement, and profitability**. 🚀

## ** 3.Do restaurants that offers table booking attract more customers**

```python
import matplotlib.pyplot as plt
import pandas as pd

df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv")

booking_analysis=df.groupby('booking')['votes'].mean()
plt.figure(figsize=(10,6))
colors=['gold' if x==max(booking_analysis) else 'royalblue' for x in booking_analysis]
booking_analysis.plot(kind='bar',color=colors,edgecolor='black')

plt.xlabel("Table Booking Available")
plt.ylabel("Average votes")
plt.title("Impact of table booking on Customer votes")
plt.xticks(rotation=0)

for i, v in enumerate(booking_analysis):
    plt.text(i, v + 20, f"{v:.1f}", ha='center', color='black', fontsize=10, fontweight='bold')

plt.savefig("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\Visualizations\\table_booking.png", dpi=300, bbox_inches='tight')
plt.show()
```

## ** Impact of Table Booking on Customer Engagement**
- **Chart Type:** Bar Chart  
- **Key Observations:**
  - Restaurants that **offer table booking** have an **average of 1,251.7 votes**.
  - Restaurants that **do not offer table booking** have an **average of only 229.4 votes**.

### 📌 **Insight:**
- Restaurants with table booking receive **more customer engagement**, as seen from the significantly higher average votes.

### ✅ **Recommendations:**
- Restaurants should **enable table booking** to attract more customers.
- Promote the table booking feature through **online platforms** and **in-store marketing**.

## ** 4.Which cities have best and worst rated restaurants**

## ** 5.Do higher votes indicates a more popular and highly rated restaurant?**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(x=df['votes'],y=df['rating'],alpha=0.5)

plt.xlabel("Number of Votes")
plt.ylabel("Restaurant Rating")
plt.title("Scatter Plot of Votes vs Rating")

plt.grid(True)
plt.savefig("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\Visualizations\\votes_vs_ratings.png", dpi=300, bbox_inches='tight')
plt.show()
```

## ** Relationship between votes vs rating**
- **Chart Type:** Scatter Plot 
- **Key Observations:**
  - More votes,better ratings **Highly-rated restaurants tend to have more votes**
  - Low votes,nixed rating **Some lesser-known restaurants still maintain high quality**
  - Votes impact stabilizes After a point, **more votes dont significantly increase rating**

### 📌 **Insight:**
- More votes often mean higher ratings,but quality and engagement matter most for reputation.
- Finally there is relationship,but its not a direct or strong **Higher votes usually mean higher ratings, but not always. Many highly-rated restaurants have a lot of votes, suggesting that popular places tend to get good ratings.** and
- **Low-vote restaurants have mixed ratings—some are excellent, some are poor. This means a restaurant with fewer votes is not necessarily bad.**
- **After a certain point, more votes don’t improve ratings much—most top-rated restaurants stabilize around 4.0 to 4.5.**
