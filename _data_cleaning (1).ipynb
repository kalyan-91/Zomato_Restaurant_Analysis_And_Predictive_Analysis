```python
import pandas as pd

try:
    df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\UncleanedData\\zomato.csv")
    print("Data loaded successfully.")
except Exception as e:
    print("Error in loading data:",e)

print(df.shape)

print(df.columns)

print(df.index)

print(df.info())

df.drop(['url','address','phone','menu_item','dish_liked','reviews_list','listed_in(city)'],axis=1,inplace=True)
df.head(5)
print(df)

print(df.info())

print(df.isnull().sum())

print(df.isnull().sum().sum())

df.rename(columns={'name':'restaurants','book_table':'booking','listed_in(city)':'city','rate':'rating','approx_cost(for two people)':'cost','listed_in(type)':'types'},inplace=True)
print(df.head(5))

print(df.dropna(inplace=True))

print(df.isnull().values.any())

print(df.info())

print(df[df.duplicated()].count().sum())

print(df.drop_duplicates(inplace=True))

print(df.shape)

print(df[df.duplicated].count().sum())

print(df['rating'].unique())

df['rating']=df['rating'].replace("NEW|-",'0',regex=True).replace('/5','',regex=True).astype("float")
print(df['rating'].head())

print(df['rating'].isnull().sum())

print(df.info())

print(df['location'].unique())

print(df['location'].isnull().sum())

a=df['location'].value_counts(ascending=False)
print(a)

print(df['cost'].unique())

df['cost']=df['cost'].replace(",",'',regex=True).astype(int)

print(df['cost'].isnull().sum())

print(df.info())

print(df['booking'].unique())

print(df['booking'].isnull().sum())

print(df['online_order'].unique())

print(df['online_order'].isnull().sum())

print(df['rest_type'].isnull().any())

print(df['rest_type'].unique())

b=df['rest_type'].value_counts(ascending=False)
print(b)

print(df['types'].unique())

c=df['types'].value_counts()
print(c)

print(df['types'].isnull().sum())

d=df['cuisines'].value_counts()
print(d)

print(df['cuisines'].unique())

print(df.groupby('restaurants').count().head())

df['restaurants']=df['restaurants'].str.replace('[Ãx][^A-Za-z]+','',regex=True)
print(df.groupby('restaurants').count().head())

print(df.info())

print(df.isnull().sum())

print(df.shape)

try:
    output_path=r"C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv"
    df.to_csv(output_path,index=False)
    print("Cleaned data successfully saved to",output_path)
except Exception as e:
    print("An error occured!",e)
```
