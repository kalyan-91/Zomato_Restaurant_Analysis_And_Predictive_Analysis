## **1. Restaurants Delivering Online or not**

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv")

delivery_counts=df['online_order'].value_counts()

plt.figure(figsize=(5,5))
colors=['skyblue','lightcoral']
plt.pie(delivery_counts,labels=delivery_counts.index,autopct='%1.1f%%',colors=colors,startangle=140)
plt.title("Online Delivery Availability")
plt.show()
```

## **2.Restaurants allowing table booking or not**

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv")
booking_counts=df['booking'].value_counts()

plt.figure(figsize=(7,7))
colors=['tomato','lightgreen']
plt.bar(booking_counts.index,booking_counts.values,color=colors)

plt.xlabel("Table Booking Availability")
plt.ylabel("Number of Restaurants")
plt.title("Restaurants Allowing Table Booking or not")

for i,v in enumerate(booking_counts.values):
    plt.text(i,v+5,str(v), ha='center', fontsize=12)

plt.show()
```

# **Visualization Report**
# **2. Table Booking Availability**
-**Chart Type:** Bar Chart
-**Key Observations:**
- Only 5386 restaurants allow table booking
- 25572 restaurants do not have table booking option
**Insight:**
  -Restaurants offering table booking tend to have higher ratings
### **Recommendations:**
- Encourage restaurants to offer **table booking** to attract high-end customers.
-  Use **peak-hour reservations** to avoid overcrowding and improve customer experience.  


## **2.Table booking rate vs rating**

```python
avg_ratings=df.groupby('booking')['rating'].mean()
plt.figure(figsize=(7,5))
colors=['purple','blue']
plt.bar(avg_ratings.index,avg_ratings.values,color=colors)
plt.xlabel("Table Booking Availablity")
plt.ylabel("Average Rating")
plt.title("Average Rating for Restaurants with and without Table Booking")

for i,v in enumerate(avg_ratings.values):
    plt.text(i,v+0.05,f"{v:2f}",ha='center',fontsize=12)
plt.show()
```

# **Visualization Report**
## **3. Table Booking Rate vs Rating**

- **Key Observations:**

- **Insight:**

### **Recommendations:**

## 7.**Most Famous Restaurants Chains in Bangalore**

```python
df['location'] = df['location'].str.strip()  # Remove any extra spaces

# Filtering locations that contain "Bangalore"
bengaluru_chains = df[df['location'].str.contains("Bangalore", case=False, na=False)]['restaurants'].value_counts().head(10)

if bengaluru_chains.empty:
    print("No restaurant chains found for Bangalore locations.")
else:
    plt.figure(figsize=(12, 6))
    colors = plt.cm.Paired.colors[:10]

    bars = bengaluru_chains.plot(kind='barh', color=colors, edgecolor='black')

    plt.xlabel("Number of Outlets")
    plt.ylabel("Restaurant Chains")
    plt.title("Top 10 Restaurant Chains in Bangalore")
    plt.xlim(0, bengaluru_chains.max() + 5)

    for bar in bars.patches:
        plt.text(bar.get_width() + 1, bar.get_y() + bar.get_height() / 2, int(bar.get_width()), 
                 va='center', color='black', fontsize=10, fontweight='bold')

    plt.gca().invert_yaxis()
    plt.savefig("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\Visualizations\\most_populat_restaurants_chains_in_banglore.png", dpi=300, bbox_inches='tight')
    plt.show()
```

## **11.Cost Vs Rating**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df=pd.read_csv("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\CleanedData\\cleaned_data.csv")
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['cost'],y=df['rating'],alpha=0.5)

sns.regplot(x=df['cost'],y=df['rating'],scatter=False,color='red')

plt.xlabel("Cost")
plt.ylabel("Rating")
plt.title("Relationship between Cost and Rating")
plt.savefig("C:\\Projects\\DataAnalysisProjects\\ZomatoRestaurantDataAnalysis\\Visualizations\\cost_and_rating.png", dpi=300, bbox_inches='tight')
plt.show()
```

 # 📊 Insight Report: Relationship Between Cost and Rating

## 🔍 Key Findings:
1. **Positive Correlation:** The scatter plot and regression line indicate a **positive relationship** between cost and rating. As restaurant cost increases, ratings also tend to be higher.  
2. **Cost Range in Scatter Plot:** The visualization shows restaurant costs ranging **up to 6000**, reflecting individual restaurant prices.  
3. **Rating Distribution:**  
   - Most restaurants with lower costs (below 2000) have varied ratings.  
   - Higher-cost restaurants generally have better ratings, indicating a **premium dining experience** may contribute to higher customer satisfaction.  

## 📌 Conclusion:
The analysis suggests that **higher-cost restaurants tend to receive better ratings**, though affordability remains a significant factor in customer choices.
